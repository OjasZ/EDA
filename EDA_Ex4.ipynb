```r
#Experiment 4: Data Cleaning and Univariate Analysis

# Data Import and Exploration
# Load necessary library
library(ggplot2)

#Read the data
co2_data <- read.csv("SuicideChina.csv")

#Check the first few rows of the dataframe
head(co2_data)

# 1.1 Basic Exploration
dim(co2_data)  # Dimensions of the dataset
names(co2_data) # Column names
str(co2_data)   # Structure of the dataset

head(co2_data)  # First few rows
tail(co2_data)  # Last few rows

# 2. Data Cleaning

#Check for missing values
print(paste("Missing values:", sum(is.na(co2_data)))) #Sum of all the missing values, will return 0 if there isn't
#if sum>0 do the following
#co2_data <- na.omit(co2_data) #Remove all the rows containing NA

# Identify numeric and factor columns
numeric_columns <- names(co2_data)[sapply(co2_data, is.numeric)]
factor_columns <- names(co2_data)[sapply(co2_data, is.factor)]

# Print the identified columns
print("Numeric Columns:")
print(numeric_columns)
print("Factor Columns:")
print(factor_columns)

# 3. Univariate Analysis
# 3.1 Summary Statistics
summary(co2_data) # Summary statistics for all columns

# 3.2 Central Tendency and Dispersion for Numeric Columns
for (col in numeric_columns) {
  if (is.numeric(co2_data[[col]])) {
    cat("Column:", col, "\n")
    cat("Mean:", mean(co2_data[[col]]), "\n")
    cat("Median:", median(co2_data[[col]]), "\n")
    cat("Variance:", var(co2_data[[col]]), "\n")
    cat("Standard Deviation:", sd(co2_data[[col]]), "\n")
    cat("Range:", range(co2_data[[col]]), "\n")
    cat("IQR:", IQR(co2_data[[col]]), "\n")
    cat("\n")
  }
}

# 3.3 Visualizations
# Function to add titles to plots
add_titles <- function(plot, title) {
  plot + labs(title = paste(title, "– 21BDS0187"),
              x = "X-axis Label",
              y = "Y-axis Label") +
    theme_minimal()
}

#Histograms
for (col in numeric_columns) {
  if (is.numeric(co2_data[[col]])) {
    hist_plot <- ggplot(co2_data, aes_string(x = col)) +
      geom_histogram(bins = 30, fill = "skyblue", color = "black")

    hist_plot <- add_titles(hist_plot, paste("Histogram of", col))
    print(hist_plot)
  }
}


#Density plots
for (col in numeric_columns) {
  if (is.numeric(co2_data[[col]])) {
    density_plot <- ggplot(co2_data, aes_string(x = col)) +
      geom_density(fill = "lightgreen", color = "black")

    density_plot <- add_titles(density_plot, paste("Density Plot of", col))
    print(density_plot)
  }
}

#Box plots
for (col in numeric_columns) {
  if (is.numeric(co2_data[[col]])) {
    box_plot <- ggplot(co2_data, aes_string(y = col)) +
      geom_boxplot(fill = "lightcoral", color = "black") +
      coord_flip()

    box_plot <- add_titles(box_plot, paste("Box Plot of", col))
    print(box_plot)
  }
}

```
