```r
#OJAS SINGH   21BDS0187

#Experiment 4: Data Cleaning and Univariate Analysis

#1. Import the Dataset
# Load necessary libraries
library(dplyr)
library(ggplot2)

# Read the dataset
df <- read.csv("SuicideChina.csv", stringsAsFactors = FALSE)

# View the first few rows
head(df)

#2. Data Cleaning
#(a) Checking Basic Information
# Dimensions of dataset (Rows and Columns)
dim(df)

# Column names
names(df)

# Structure of dataset
str(df)

# Display first and last few rows
head(df)
tail(df)

# Summary of dataset
summary(df)

#(b) Handling Missing Values
# Count missing values in each column
colSums(is.na(df))

# Remove rows with missing values (if needed)
df_cleaned <- na.omit(df)

# Fill missing numerical values with mean
df$Age[is.na(df$Age)] <- mean(df$Age, na.rm = TRUE)

# Fill missing categorical values with mode
fill_mode <- function(x) {
  mode_value <- names(sort(table(x), decreasing = TRUE))[1]
  x[is.na(x)] <- mode_value
  return(x)
}
df$Education <- fill_mode(df$Education)
df$Occupation <- fill_mode(df$Occupation)

#3. Univariate Analysis
#(a) Measure of Central Tendency
# Mean, Median, Mode of Age
mean(df$Age, na.rm = TRUE)
median(df$Age, na.rm = TRUE)

# Mode function
mode_function <- function(x) {
  return(names(sort(table(x), decreasing = TRUE))[1])
}
mode_function(df$Age)

#(b) Measure of Dispersion
# Standard Deviation and Variance of Age
sd(df$Age, na.rm = TRUE)
var(df$Age, na.rm = TRUE)

# Interquartile Range (IQR) of Age
IQR(df$Age, na.rm = TRUE)

# Range of Age
range(df$Age, na.rm = TRUE)


#(c) Correlation Analysis
## Correlation between numerical columns
cor(df[c("Age", "Year")], use = "complete.obs")


#4. Data Visualization
#(a) Histogram of Age Distribution
ggplot(df, aes(x = Age)) +
  geom_histogram(binwidth = 5, fill = "blue", color = "black") +
  labs(title = "Age Distribution – 23BDS0187", x = "Age", y = "Frequency")


#(b) Density Plot of Age
ggplot(df, aes(x = Age)) +
  geom_density(fill = "blue", alpha = 0.5) +
  labs(title = "Density Plot of Age – 23BDS0187", x = "Age", y = "Density")


#(c) Histogram and Density Plot Together
ggplot(df, aes(x = Age)) +
  geom_histogram(aes(y = ..density..), binwidth = 5, fill = "blue", color = "black", alpha = 0.6) +
  geom_density(color = "red", size = 1) +
  labs(title = "Age Distribution with Density – 23BDS0187", x = "Age", y = "Density")


#(d) Boxplot of Age by Sex
ggplot(df, aes(x = Sex, y = Age, fill = Sex)) +
  geom_boxplot() +
  labs(title = "Age Distribution by Sex – 23BDS0187", x = "Sex", y = "Age")


#(e) Pie Chart of Suicide Methods
## Count occurrences of each suicide method
method_counts <- table(df$method)

# Create pie chart
pie(method_counts, main = "Suicide Methods – 23BDS0187", col = rainbow(length(method_counts)))


#(f) Dot Plot of Age
ggplot(df, aes(x = Age)) +
  geom_dotplot(binwidth = 5, fill = "blue") +
  labs(title = "Dot Plot of Age – 23BDS0187", x = "Age", y = "Count")


#5. Saving the Cleaned Data
write.csv(df_cleaned, "SuicideChina_Cleaned.csv", row.names = FALSE)
```
